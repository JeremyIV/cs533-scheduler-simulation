```python
import numpy as np
import matplotlib.pyplot as plt
from main import execute_simulations
```

```python
turnarounds = {}
responses = {}

for scheduler_type in ['RR', 'random', 'FIFO', 'weightedRandom']:
    turnaround, response = execute_simulations(scheduler_type, 100, 1000, 1000)
    turnarounds[scheduler_type] = turnaround
    responses[scheduler_type] = response
```

```python
log_turnarounds = {k: np.log10(t) for k, t in turnarounds.items()}
log_responses = {k: np.log10(t) for k, t in responses.items()}
```

```python
all_turnarounds = np.concatenate(list(log_turnarounds.values()))
bins = np.linspace(all_turnarounds.min(), all_turnarounds.max(), 30)
for scheduler_type in log_turnarounds.keys():
    plt.hist(log_turnarounds[scheduler_type], bins=bins, label=scheduler_type, alpha=0.4)
plt.legend()
plt.title('mean turnaround time')
plt.xlabel('log(mean turnaround time)')
plt.ylabel('frequency')
```

```python
all_responses = np.concatenate(list(log_responses.values()))
bins = np.linspace(all_responses.min(), all_responses.max(), 30)
for scheduler_type in log_responses.keys():
    plt.hist(log_responses[scheduler_type], bins=bins, label=scheduler_type, alpha=0.4)
plt.legend()
plt.title('mean response time')
plt.xlabel('log(mean response time)')
plt.ylabel('frequency')
```
